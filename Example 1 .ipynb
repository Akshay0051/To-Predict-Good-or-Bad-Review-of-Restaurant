```python
# text cleaning 

import string 

def text_cleaning(text):
    
    # remove punchuation 
    clean = []
    for char in text:
        if char not in string.punctuation:
            clean.append(char)
            
    clean = " ".join(clean)
    
      
    cleaning = []
    
    for word in clean.split():
        if word.lower() not in stopwords.words('english'):
            cleaning.append(word.lower())
    
            
    return cleaning
```

```python
data = text_cleaning(X)
data
```
